```python
import pandas as pd
import zipfile
```

```python
zip_path = 'RiceExporters.zip'
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall('extracted_folder')

# List all CSV files in the extracted folder
csv_files = [file for file in zip_ref.namelist() if file.endswith('.csv')]

# Initialize an empty list to hold DataFrames
dataframes = []

# Read and concatenate all CSV files into a single DataFrame
for file in csv_files:
    df = pd.read_csv(f'extracted_folder/{file}')
    dataframes.append(df)

# Concatenate all DataFrames into a single DataFrame
combined_data = pd.concat(dataframes, ignore_index=True)
```

```python
df = combined_data
```

```python
# Define the columns to keep
columns_to_keep = [
    'NO',
    'ARRIVAL DATE',
    'HS CODE',
    'HS CODE DESCRIPTION',
    'IMPORTER NAME',
    'IMPORTER COUNTRY',
    'IMPORT VALUE CIF',
    'CITY STATE',    
    'EXPORTER NAME',
    'EXPORTER ADDRESS',
    'COUNTRY OF ORIGIN',
    'IMPORT VALUE FOB',    
    'CURRENCY',    
    'QUANTITY',
    'QUANTITY UNIT',
    'PRODUCT DETAILS',    
    'PORT OF ARRIVAL',
    'PORT OF DEPARTURE'
    
]

# Keep only the specified columns in the DataFrame
data = df[columns_to_keep]
```

```python
data['QUANTITY UNIT'].unique()
```

```python
mapping = {
    'KILOGRAM': 'KGS',
    'KILOGRAMOS NETOS': 'KGS',
    'TON': 'MTS',
    'TNE': 'MTS'
}

# Replace values in 'QUANTITY UNIT' column based on the mapping
data['QUANTITY UNIT'] = data['QUANTITY UNIT'].replace(mapping)

# Check the unique values after renaming
print(data['QUANTITY UNIT'].unique())
```

```python
# Define the values to filter
values_to_filter = ['KGS', 'MTS', 'LBS', 'QTL']

# Filter the DataFrame based on specified values in 'QUANTITY UNIT' column
filtered_data = data[data['QUANTITY UNIT'].isin(values_to_filter)]
```

```python
data = filtered_data
data['QUANTITY UNIT'].unique()
```

```python
# Define a function to calculate QTY IN KG
def calculate_qty(row):
    if row['QUANTITY UNIT'] == 'MTS':
        return row['QUANTITY'] * 1000
    elif row['QUANTITY UNIT'] == 'LBS':
        return row['QUANTITY'] * 0.453
    if row['QUANTITY UNIT'] == 'QTL':
        return row['QUANTITY'] * 100
    else:
        return row['QUANTITY']

# Apply the function to create the new column 'QTY IN KG'
data['QTY IN KG'] = data.apply(calculate_qty, axis=1)
```

```python
Q1 = data['QTY IN KG'].quantile(0.25)
Q3 = data['QTY IN KG'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define Outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Outliers
outliers = data[(data['QTY IN KG'] < lower_bound) | (data['QTY IN KG'] > upper_bound)]

print(outliers)
```

```python
data_clean = data[(data['QTY IN KG'] >= lower_bound) & (data['QTY IN KG'] <= upper_bound)]
data = data_clean
data.head()
```
